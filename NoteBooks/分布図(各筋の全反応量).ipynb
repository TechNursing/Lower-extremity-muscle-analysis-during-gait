```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

n_rows = 5
n_cols = 2
fig, axes = plt.subplots(n_rows, n_cols * 2, figsize=(20, 25))
plt.subplots_adjust(hspace=0.4, wspace=0.4)

for i, data in enumerate(data_list):
    row = i // 2
    col = (i % 2) * 2
    ax_hist = axes[row, col]
    ax_qq = axes[row, col + 1]

    # Tibialis anteliorのデータ
    data_tibialis = data['Tibialis antelior']
    mu = data_tibialis.mean()
    sigma = data_tibialis.std()

    ax_hist.hist(data_tibialis, bins=30, density=True, alpha=0.6, color='g')
    xmin, xmax = ax_hist.get_xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, mu, sigma)
    ax_hist.plot(x, p, 'k', linewidth=2)

    if i < len(data_list) - 1:
        title_hist = f"被験者{i+1}の前脛骨筋の正規分布との近似\n母平均 = {mu:.2f}, 標準偏差 = {sigma:.2f}"
        title_qq = f"Q-Qプロット\n(被験者{i+1}の前脛骨筋の正規分布との近似)"
    else:
        title_hist = f"被験者全員の前脛骨筋の正規分布との近似\n母平均 = {mu:.2f}, 標準偏差 = {sigma:.2f}"
        title_qq = f"Q-Qプロット\n(被験者全員の前脛骨筋の正規分布との近似)"

    ax_hist.set_title(title_hist)
    stats.probplot(data_tibialis, dist="norm", plot=ax_qq)
    ax_qq.set_title(title_qq)

for j in range(i + 1, n_rows * n_cols):
    fig.delaxes(axes[j // 2, (j % 2) * 2])
    fig.delaxes(axes[j // 2, (j % 2) * 2 + 1])

results_path = '../Plots/各被験者と全員の前脛骨筋の正規分布との近似とQ-Qプロット.png'
plt.savefig(results_path, bbox_inches='tight')
plt.show()
```

```python
def save_text_image(texts, filename):
    fig, ax = plt.subplots(figsize=(10, len(texts) * 0.3))
    for i, text in enumerate(texts):
        ax.text(0.5, 1 - (i + 1) / len(texts), text, fontsize=12, ha='center', va='center', wrap=True)
    ax.axis('off')
    plt.savefig(filename, bbox_inches='tight', pad_inches=0.1)
    plt.close(fig)

results = []
for i, data in enumerate(data_list, start=1):
    data_tibialis = data['Tibialis antelior']
    mu = np.mean(data_tibialis)
    sigma = np.std(data_tibialis)
    
    kstest_result = stats.kstest(data_tibialis, 'norm', args=(mu, sigma))
    ks_result = f"被験者{i}のコルモゴロフ=スミルノフ検定の結果: 統計量 = {kstest_result.statistic:.4f}, p値 = {kstest_result.pvalue:.4f}"
    
    adtest_result = stats.anderson(data_tibialis, dist='norm')
    ad_result = f"被験者{i}のアンダーソンｰダーリング検定の結果: 統計量 = {adtest_result.statistic:.4f}"
    
    if i < len(data_list):
        results.append(ks_result)
        results.append(ad_result)
        for significance_level, critical_value in zip(adtest_result.significance_level, adtest_result.critical_values):
            results.append(f"    有意水準 = {significance_level}%, 臨界値 = {critical_value:.4f}")
        results.append("")
    else:
        ks_result = f"被験者全員のコルモゴロフ=スミルノフ検定の結果: 統計量 = {kstest_result.statistic:.4f}, p値 = {kstest_result.pvalue:.4f}"
        ad_result = f"被験者全員のアンダーソンｰダーリング検定の結果: 統計量 = {adtest_result.statistic:.4f}"
        results.append(ks_result)
        results.append(ad_result)
        for significance_level, critical_value in zip(adtest_result.significance_level, adtest_result.critical_values):
            results.append(f"    有意水準 = {significance_level}%, 臨界値 = {critical_value:.4f}")
        results.append("")

for result in results:
    print(result)

save_text_image(results, '../Plots/正規分布の検定(KS&AD).png')
```

```python
import warnings

def fit_and_plot_distribution(data, dist_name, ax, i, is_last):
    dist = getattr(stats, dist_name)
    try:
        with warnings.catch_warnings():
            warnings.filterwarnings('ignore')
            params = dist.fit(data['Tibialis antelior'])
        ax.hist(data['Tibialis antelior'], bins=30, density=True, alpha=0.6, color='g')
        xmin, xmax = ax.get_xlim()
        x = np.linspace(xmin, xmax, 100)
        p = dist.pdf(x, *params)
        dist_name_japanese = {
            'lognorm': '対数正規分布',
            'expon': '指数分布',
            'gamma': 'ガンマ分布',
            'beta': 'ベータ分布',
            'weibull_min': 'ワイブル分布'
        }.get(dist_name, dist_name)
        
        title = f'被験者全員の{dist_name_japanese}のフィット' if is_last else f'被験者{i+1}の{dist_name_japanese}のフィット'
        ax.plot(x, p, 'k', linewidth=2)
        ax.set_title(title)
    except Exception as e:
        print(f"被験者{i+1}の{dist_name}のフィットに失敗しました: {e}")
        ax.set_title(f'被験者{i+1}の{dist_name_japanese}のフィットに失敗')

n_rows = len(data_list)
n_cols = 5  
fig, axes = plt.subplots(n_rows, n_cols, figsize=(25, n_rows * 5))
fig.subplots_adjust(hspace=0.4)

for i, data in enumerate(data_list):
    is_last = (i == len(data_list) - 1)
    fit_and_plot_distribution(data, 'lognorm', axes[i, 0], i, is_last)
    fit_and_plot_distribution(data, 'expon', axes[i, 1], i, is_last)
    fit_and_plot_distribution(data, 'gamma', axes[i, 2], i, is_last)
    fit_and_plot_distribution(data, 'beta', axes[i, 3], i, is_last)
    fit_and_plot_distribution(data, 'weibull_min', axes[i, 4], i, is_last)

results_path = '../Plots/各分布図.png'
plt.savefig(results_path, bbox_inches='tight')
plt.show()
```

```python
def plot_qq(data, dist_name, ax, is_last, i):
    dist = getattr(stats, dist_name)
    try:
        with warnings.catch_warnings():
            warnings.filterwarnings('ignore')
            params = dist.fit(data['Tibialis antelior'])
        stats.probplot(data['Tibialis antelior'], dist=dist_name, sparams=params, plot=ax)
        dist_name_japanese = {
            'lognorm': '対数正規分布',
            'expon': '指数分布',
            'gamma': 'ガンマ分布',
            'beta': 'ベータ分布',
            'weibull_min': 'ワイブル分布'
        }.get(dist_name, dist_name)
        title = f'被験者全員の{dist_name_japanese}のQ-Qプロット' if is_last else f'被験者{i+1}の{dist_name_japanese}のQ-Qプロット'
        ax.set_title(title)
    except Exception as e:
        print(f"被験者{i+1}の{dist_name}のフィットに失敗しました: {e}")
        ax.set_title(f'被験者{i+1}の{dist_name_japanese}のQ-Qプロットに失敗')

def test_distribution(data, dist_name, params):
    data_values = data['Tibialis antelior']
    ks_result = stats.kstest(data_values, dist_name, args=params)
    return ks_result

n_rows = len(data_list)
n_cols = 5  
fig, axes = plt.subplots(n_rows, n_cols, figsize=(25, n_rows * 5))
fig.subplots_adjust(hspace=0.4)

results = []

dist_names = ['lognorm', 'expon', 'gamma', 'beta', 'weibull_min']

for i, data in enumerate(data_list):
    is_last = (i == len(data_list) - 1)
    
    for j, dist_name in enumerate(dist_names):
        ax = axes[i, j]
        plot_qq(data, dist_name, ax, is_last, i)
        try:
            with warnings.catch_warnings():
                warnings.filterwarnings('ignore')
                params = stats.__dict__[dist_name].fit(data['Tibialis antelior'])
            ks_result = test_distribution(data, dist_name, params)
        except Exception as e:
            print(f"被験者{i+1}の{dist_name}のフィットに失敗しました: {e}")
            ks_result = None
        
        dist_name_japanese = {
            'lognorm': '対数正規分布',
            'expon': '指数分布',
            'gamma': 'ガンマ分布',
            'beta': 'ベータ分布',
            'weibull_min': 'ワイブル分布'
        }.get(dist_name, dist_name)
        
        if ks_result is not None:
            if is_last:
                results.append(f"被験者全員の{dist_name_japanese}のコルモゴロフ=スミルノフ検定: 統計量 = {ks_result.statistic:.4f}, p値 = {ks_result.pvalue:.4f}")
            else:
                results.append(f"被験者{i+1}の{dist_name_japanese}のコルモゴロフ=スミルノフ検定: 統計量 = {ks_result.statistic:.4f}, p値 = {ks_result.pvalue:.4f}")

results_path = '../Plots/各分布のQQプロット.png'
plt.savefig(results_path, bbox_inches='tight')
plt.show()

for result in results:
    print(result)

save_text_image(results, '../Plots/各分布のコルモゴロフ=スミルノフ検定.png')
```
